```python
#Multivariate Regression
#importing relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
#DataCleaning_using interpolation
missing_values=["?"]
names = ['mpg','cylinders','displacement','horsepower','weight','accelaration','model_year','origin']
df=pd.read_csv("auto-mpg.csv", names=names,na_values=missing_values) 
#df = (df-df.mean())/df.std()
df
```

```python
xx=df.interpolate()
xx
```

```python
inputx1 = xx.iloc[:,1:-2].values # cylinders to acceleration
outputy1 = xx.iloc[:,0].values  # mpg
print(inputx1)
print(outputy1)
```

```python
input_train1, input_test1, output_train1, output_test1 = train_test_split(inputx1, outputy1, test_size = 1/5, random_state = 1)
print(input_train1)
print(input_test1)  
print(output_train1)
print(output_test1)
```

```python
model = LinearRegression()
model.fit(input_train1, output_train1)
predicted_output = model.predict(input_test1)
print(predicted_output)
cylinders=float(input("cylinders->"))
displacement=float(input("displacement->"))
Horsepower=float(input("Horsepower->"))
weight=float(input("weight->"))
acceleration=float(input("acceleration->"))
testinput = [[cylinders,displacement,Horsepower,weight,acceleration]]
predicted_output = model.predict(testinput)
print("Predicted Mpg->{0}".format(predicted_output))
```

```python
missing_values=["?"]
names = ['mpg','cylinders','displacement','horsepower','weight','accelaration','model_year','origin']
df=pd.read_csv("auto-mpg.csv", names=names,na_values=missing_values) 
xx=df.interpolate() # removing redundant values by interpolation

xx = (xx - xx.mean())/xx.std() # normalizing

X = xx.iloc[:,1:5] 

ones = np.ones([X.shape[0],1])
X = np.concatenate((ones,X),axis=1)
#print(X)
y = xx.iloc[:,0:1].values #.values converts it from pandas.core.frame.DataFrame to numpy.ndarray
#print(y)
theta = np.zeros([1,5])

#cost function
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))

#using gradient descent method
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost

#set hyper parameters
alpha = 0.0001
iters = 1000

g,cost = gradientDescent(X,y,theta,iters,alpha)
print(g)
print(np.shape(g))
finalCost = computeCost(X,y,g)
print(finalCost)

alpha = 0.0001
iters = 1000


fig, ax = plt.subplots()  
#ax.scatter(theta,y , color = 'green')
ax.plot(np.arange(iters), cost, 'r')  
ax.set_xlabel('Iterations')  
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Training Iterations')
plt.show()
```
